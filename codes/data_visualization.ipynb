```python
##TODO
# - change the mjdb time to proper time class
# - create a new datraframe like train numpy and valid numpy 
# - load the entire dataset and save theat explcitily

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.debugger import Tracer

object_ids={}; brightness={}

## meta data
training_meta = pd.read_csv('../data/PLAsTiCC-2018/training_set_metadata.csv')
target_values = list(set(training_meta.target))
for target in target_values:
    object_ids[target] = list(set(training_meta[training_meta['target']==target].object_id))
    print(f'target = {target} len of object_ids = {len(object_ids[target])}')

## reading sample data
test_set_sample = pd.read_csv('../data/PLAsTiCC-2018/test_set_sample.csv')
test_set_sample.head()

## combining through object id 
for target in target_values:
    target_brightness = test_set_sample[test_set_sample['object_id']==target]
    if not target_brightness.empty:
        brightness[target]={}
        for passband in set(target_brightness.passband):
            brightness[target][passband] = target_brightness[target_brightness['passband']==passband].flux

## plotting for a particular traget with all passbands
target1 = 88
for passband in brightness[target1].keys():
    plt.plot(brightness[target1][passband])
plt.show()
```
